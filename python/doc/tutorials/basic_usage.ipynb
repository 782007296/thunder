# Basic usage

thunder offers a variety of analyses and workflows for spatial and temporal data. We'll walk through a very simple example here as an introduction. The entry point for most workflows is the ``ThunderContext``. If you type ``thunder`` to start the interactive shell, this context is automatically provided as ``tsc``, which is a thin wrapper for the ``SparkContext`` that provides functionality for loading data.

We'll start by loading some example data

```python
data = tsc.loadExample('fish')
```

``data`` is a distributed collection of key-value records, each containing a coordinate identifier and the time series of a single pixel. We can look at the first record by using ``first()``

```python
data.first()
```

We can compute the mean value for each pixel using the ``seriesMean`` method

```python
means = data.seriesMean()
means.first()
```

``means`` contains the mean of each pixel, which we can ``pack`` into a local image using

```python
img = means.pack(sorting=True)
```

It's easy to look at the result directly using ``matplotlib``. We'll use the excellent ``seaborn`` package for styling figures, but this is optional.

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("notebook")
sns.set_style("white")
plt.imshow(img);
```

It's also easy to save the result to a ``numpy`` file, or to a ``MAT`` file if you do the rest of your analysis in Matlab.

```python
from thunder.utils import export
export(img, "directory", "meanval", "npy")
export(img, "directory", "meanval", "matlab")
```

This will put a ``numpy`` file or ``MAT`` file called ``meanval`` in the folder ``directory`` in your current directory.
