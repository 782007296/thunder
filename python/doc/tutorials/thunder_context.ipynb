# Thunder context

The `ThunderContext` is the entry point for most analyses. It is automatically provided as tsc when you start the interactive shell using ``thunder``. It can also be created manually, in two different ways. These are useful when writing standalone analysis scripts. First, it can be created from an existing instance of a ``SparkContext``:

```python
from thunder.utils.context import ThunderContext
tsc = ThunderContext(sc)
```

Or it can be created directly using the same arguments provided to a ``SparkContext``:

```python
tsc = ThunderContext.start(appName='myapp')
```

The primary methods for loading data are `loadSeries` and `loadImages`, for loading a `Series` or `Images` object, respectively. See the Input Format tutorial for more information on loading and data types.

The `makeExample` method makes it easy to generate example data for testing purposes, by calling methods from the `DataSets` class:

```python
data = tsc.makeExample('kmeans', k=2, ndims=10, nrecords=10, noise=0.5)
```

```python
%matplotlib inline
```

```python
from numpy import asarray
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('notebook')
ts = data.values().collect()
plt.plot(asarray(ts).T);
```

The 'loadExample' method loads one of the local example datasets. This are highly compressed and downsampled, and meant only to demonstrate basic functionality, not to represent anything meaningful about the data itself.

```python
data = tsc.loadExample('fish')
```

```python
img = data.seriesMean().pack(sorting=True)
```

```python
sns.set_style('white')
sns.set_context('notebook')
plt.imshow(img);
```

Example data stored on S3 can be loaded using the :func:'loadExampleEC2' method:

```python
data, params = tsc.loadExampleEC2('zebrafish-optomotor-response')
```

You must be running thunder on an Amazon EC2 cluster in order to use these methods. Some of the available data sets include parameters that are returnded alongside the data.
