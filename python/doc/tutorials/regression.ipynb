# Regression

Linear regression is a simple and powerful tool for describing relationships between a response variable and one or more explanatory variables. Here, we show how to peform ordinary least squres regression in Thunder and visualize the results.

Load and convert data
---

Load a series of images and cache the data to speed up subsequent operations.

```python
data = tsc.loadExample('mouse-images')
data.cache();
```

Take a look at the first image.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('notebook')

sns.set_style('white')
plt.imshow(data.first()[1], cmap='Blues');
```

Convert from Images to TimeSeries and cache results.

```python
series = data.toSeries()
series.cache();
```

Take a look at the time-series of a few pixels (selct ones with large standard devaitions to avoid pixels with no activity).

```python
import numpy as np
sample = series.filterOnValues(lambda x: np.std(x)>125).values().collect()

sns.set_style('darkgrid')
plt.plot(np.array(sample).T);
```

Load in associated behavioral data to be used as explanatory variables (a.k.a regressors) in predicting the above time-series.

```python
regressors = tsc.loadExample('mouse-params')
print regressors.names()
```

Correlation analysis
---

The correlation coefficient, provides a measure of the strength of the linear relationship between two variables. When we have only a single explanatory varaible, the correlation coefficient offers a simpler alternative to a regression analysis. Let's start by correlating the activity from each pixel with one of our explanatory variables.

```python
corrs = series.correlate(regressors['runningSpeed'])
```

Now we can pack these correlation coefficients back into an arrary and visualize the results. We center the values of the correlation coefficients so that positive correlations are red and negative correlations are blue.

```python
def centerForColoring(x):
    y = x[:]
    y[y>0] = y[y>0]/np.max(y[y>0])
    y[y<0] = y[y<0]/-np.min(y[y<0])
    return y

corrMat = centerForColoring(corrs.pack())

sns.set_style('white')
plt.imshow(corrMat, cmap='coolwarm');
plt.colorbar();
```

Regression analysis
---

With more than two explanatory variables, correlation is no longer a viable optoin, so we turn to linear regression. Linear regression in Thunder models the data as:

$y=\beta X$

where:
+ $y$ is a row-vector of measurements of the response variable
+ $\beta$ is a row-vector of regression coeffients, one for each regressor
+ $X$ is a design matrix with one row for each regressor and one column for each measurement

Since our data are time series, the columns of $y$ and $X$ represent measurements at different instants in time. Note: Thunder automatically adds a row of 1's to the design matrix to model an intercept term.

```python
X = regressors[['runningSpeed', 'corridorPosition']]
```

Unlike the correlation coefficient, linear regression is sensitive to the scale of the regressor. Thus, we standardize our regressors so that the resulting regression coefficients can be compared on the same scale. This will be useful later for visualization of the results.

```python
from scipy.stats.mstats import zscore
X = zscore(X, axis=1)
```

For each pixel/series fit a linear regression model that predicts the pixel values from our regressors

```python
from thunder import RegressionModel
results = RegressionModel.load(X, 'linear').fit(series)
```

The result is a Series with three indices:
+ "betas": the regression coefficients
+ "stats": $R^2$ value
+ "resid": the residuals

We can collect the regression coefficients and visualize how they are distribution across the image.

```python
b1, b2 = results.select('betas').pack()
rsq = results.select('stats').pack()
```

```python
b1map, b2map = centerForColoring(b1), centerForColoring(b2)

plt.subplots(1, 3, figsize=(15 ,10))
plt.subplot(1, 3, 1)
plt.imshow(b1map, cmap='seismic')
plt.title('b1');
plt.subplot(1, 3, 2)
plt.imshow(b2map, cmap='seismic');
plt.title('b2');
plt.subplot(1, 3, 3)
plt.imshow(rsq, interpolation='none', cmap='bone_r');
plt.title('r-squared');
```

While informative, it is hard to integrate/compare the information distributed across these three images. An alternative is to make a single map that encodes one or more of these pieces of information. One option is to express the vector of the two regression coefficients, $(\beta_1,\,\beta_2)$, in polar coordiantes. Now the angle, which represents the relative contributions of the coefficients, can be maped to color (hue in HSV) while the magnitude can be mapped to brightness (saturation in HSV). This can be accomplished with the "polar" setting on Colorize.

```python
from thunder import Colorize
maps = Colorize(toType='polar', scale=5).images(np.array([b1,b2]))
plt.imshow(maps);
```

One draw-back of of the above map is that it does not include information about $R^2$, so not all of the pixels necessarily represent reliable trends. To address this, we can apply a mask to remove all pixels where $R^2$ is less than some threshold.

```python
maps = Colorize(toType='polar', scale=5).images(np.array([b1,b2]), mask=rsq>0.1)
plt.imshow(maps);
```
