# Regression

Linear regression is a simple and powerful tool for describing relationships between a response variable and one or more explanatory variables. Here, we show how to peform ordinary least squres regression in Thunder and visualize the results.

Load and convert data
---

Load a series of images and cache to speed up subsequent operations.

```python
data = tsc.loadExample('mouse-images')
data.cache();
```

Take a look at the first image.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('notebook')

sns.set_style('white')
plt.imshow(data.first()[1], cmap='Blues');
```

Convert from Images to TimeSeries and cache results.

```python
series = data.toSeries()
series.cache();
```

Take a look at the time-series of a few pixels (selct ones with large standard devaitions to avoid pixels with no activity).

```python
import numpy as np
sample = series.filterOnValues(lambda x: np.std(x)>125).values().collect()


sns.set_style('darkgrid')
plt.plot(np.array(sample).T);
```

Load in associated behavioral data to be used as explanatory variables (a.k.a regressors) in predicting the above time-series.

```python
regressors = tsc.loadExample('mouse-params')
print regressors.names()
```

Correlation analysis
---

When we only have a single explanatory variable, the correlation coefficient - which provides a measure of the strength of the linear relationship between this variable and the response - offers a simpler alternative to a regression analysis. Let's start by correlating the activity from each pixel with one of our explanatory variables.

```python
corrs = series.correlate(regressors['runningSpeed'])
```

Now we can pack these correlation coefficients back into an arrary representing pixel location and visualize the results

```python
corrMat = corrs.pack()

sns.set_style('white')
plt.imshow(corrMat, cmap='coolwarm');
```

Regression analysis
---

With more than two explanatory variables, correlation is not an viable analysis, so we turn to linear regression. Linear regression in Thunder models the data as:

$y=\beta X$

where:
+ $y$ is a row-vector of measurements of the response variable
+ $\beta$ is a row-vector of regression coeffients, one for each regressor
+ $X$ is a design matrix one row for each regressor and a column for each measurement

Since our data are time series, each columnt of $y$ and $X$ represent measurements at one instant in time. Note: Thunder automatically adds a row of 1's to the design matrix to model an intercept term.

```python
X = np.vstack((regressors['runningSpeed'], regressors['corridorPosition']))
```

Unlike the correlation coefficient, linear regression is sensitive to the scale of the regressor. Thus, we standardize our regressors so that the resulting regression coefficients can be compared on the same scale. This will be useful later for visualization of our results

```python
from scipy.stats.mstats import zscore
X = zscore(X, axis=1)
```

For each pixel/series fit a linear regression model that predicts the pixel values from our regressors

```python
from thunder import RegressionModel
results = RegressionModel.load(X, 'linear').fit(series)
```

The result is a Series with three elements:
+ "betas": the regression coefficients
+ "stats": $R^2$ value
+ "redid": the residuals

We can collect the regression coefficients and visualize how they are distribution across the image space.

```python
b1, b2 = results.select('betas').pack()
rsq = results.select('stats').pack()
```

```python
plt.subplots(1, 3, figsize=(20,10))
plt.subplot(1, 3, 1)
plt.imshow(b1, interpolation='none', cmap='Blues');
plt.subplot(1, 3, 2)
plt.imshow(b2, interpolation='none', cmap='Greens');
plt.subplot(1, 3, 3)
plt.imshow(rsq, interpolation='none', cmap='Purples');
```

While informative, it is hard to integrate/compare the information distributed across these three images. An alternative is to make a single map that encodes all three pieces of information. We do this by converting $(\beta_1,\,\beta_2)$ to polar coordinates and mapping the angle to hue and the magnitude to saturation. Finally, the $R^2$ value is mapped to value. Doing this reveals:
+ black: many pixles show no linear relationship
+ 

```python
from matplotlib.colors import hsv_to_rgb

radius = np.sqrt(np.square(b1) + np.square(b2))
theta = np.arctan2(b2, b1)
height = rsq

r = (radius-np.min(radius))/(np.max(radius)-np.min(radius))
#r = radius/np.max(radius)
th = (theta + np.pi)/(2*np.pi)
h = (height-np.min(height))/(np.max(height)-np.min(height))
#h = height/np.max(height)

mat = np.vstack((np.ravel(th), np.ravel(r), np.ravel(h))).T
colors = hsv_to_rgb(mat)

plt.imshow(np.reshape(colors, (64, 64, 3)), interpolation='none');
```
