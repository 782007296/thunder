# Input formats

Data can be loaded from a variety of sources, depending on the desired data type.

Loading Images
--------------
Images data are distributed collections of images or volumes. Two-dimensional images can be loaded from `png` or `tif` files, and three-dimensional volumes can be loaded from multi-page `tif-stack` volumes, or binary `stack` volumes. For example, to load a set of `tif` images just by specifying a folder:

```python
data = tsc.loadImages('/Users/freemanj11/fish/tif/', inputformat='tif-stack')
```

To look at the first image:

```python
%matplotlib inline
```

```python
from numpy import amax
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
sns.set_context('notebook')
img = data.first()[1]
plt.imshow(img[:,:,0]);
```

To check the dimensions of the images:

```python
data.dims
```

And to check the total number of images:

```python
data.nimages
```

For any of these image formats, the raw data can be stored on a local file system, a networked file system (accessible to all nodes of a cluster), or Amazon S3.

It's easy to load only a subset of images: 

```python
data = tsc.loadImages('/Users/freemanj11/fish/tif/', inputformat='tif-stack', startidx=0, stopidx=10)
```

```python
data.nimages
```

Loading Series
--------------
A ``Series`` object is a distributed collections of one-dimensional arrays that share a common index, and have simple key identifiers. They can be loaded from flat text or binary files. 

Text files must contain a line for each record, with numbers separated by spaces. The first numbers of each line will be interpreted as keys, and subsequent numbers will be interperted as values. The number of keys is user-specified.

```python
data = tsc.loadSeries('/Users/freemanj11/github/thunder/python/thunder/utils/data/iris.txt', inputformat='text', nkeys=3)
```

```python
data.first()
```

For comparison, you can look at the raw lines of the text file:

```python
data = sc.textFile('/Users/freemanj11/github/thunder/python/thunder/utils/data/iris.txt')
data.first()
```

Flat binary files must store each record as a contiguous sequence of bytes, with a fixed size in bytes for each record, including both keys and values. The number and numerical type of keys and records are most convieniently specified in a configuration file in the same directory as the data, but can also be specified as input arguments.

```python
data = tsc.loadSeries('/Users/freemanj11/fish/binary/', inputformat='binary')
```

```python
data.first()
```

Both text and binary data can be loaded from a single file or multiple files, stored on a local file system, a networked file system, Amazon S3, or HDFS. To load multiple files at once, specify a directory as the filename, or a wildcard pattern.
